```python
%matplotlib inline
```

```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
#set x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

#set the y-axis to a list of floatss as the total fare in US $ accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Line Chart

### MATLAB Approach

```python
#create the plot and add label for the legend
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')
#create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
#set y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("PyBer Fare By Month")
#add a grid
plt.grid()
#add legend
plt.legend()
```

### Object Approach

```python
#create the plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, 'dg-', label = 'Boston', linewidth = 2)
#setting title and axes labels
ax.set_title("PyBer Fare By Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#setting y-axis limit
ax.set_ylim(0,45)
#adding gridlines
ax.grid()
#adding legend
ax.legend()
```

## Bar Chart

### MATLAB Approach

#### Vertical Bar Chart

```python
#create the plot 
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')
#create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create title
plt.title('PyBer Fare by Month')
#add legend
plt.legend()
```

#### Horizontal Bar Chart

```python
#create the plot
plt.barh(x_axis, y_axis, color = 'magenta', label = 'Boston')
#invert y axis
plt.gca().invert_yaxis()
#add legend
plt.legend()
#add titles and axis labels
plt.title("PyBer Fare By Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

### Object Approach

#### Vertical Chart

```python
#create plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#create plot
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label = 'Chicago', color = 'cyan')
#add title and labels
ax.set_title("PyBer Fare By Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
#adding legend
ax.legend()
#inverting axes
plt.gca().invert_yaxis()
```

## Scatter Plots and Bubble Charts

### MATLAB Approach

#### Scatter Plot

##### plt.plot()

```python
#create plot
plt.plot(x_axis, y_axis, 'o')
```

##### plt.scatter()

```python
#create plot
plt.scatter(y_axis, x_axis, color = 'red', label = 'Chicago')
#add legend
plt.legend()
#tile and axis
plt.title("PyBer Fare By Month")
plt.ylabel("Fare($)")
plt.xlabel("Date")
#add limit to x-axis data
plt.xlim(0,50)
#invert axis so Jan is on top
plt.gca().invert_yaxis()
```

#### Bubble Chart

```python
#create plot
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

### Object Approach

#### Scatter Plot

```python
#create plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

#### Bubble Chart

```python
#create plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color = 'skyblue', edgecolors = 'black', linewidth = 2, alpha = .80, label = 'Boston')
#adding legend
ax.legend()
#adding title and labels
ax.set_title("PyBer Fare By Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
#adding limit for x axis
ax.set_xlim(0,50)
#inverting y-axis
plt.gca().invert_yaxis()
```

## Pie Charts

### MATLAB Approach

```python
#adding 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#explode
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#create a plot
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode = explode_values, colors = colors, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

### Object Approach

```python
fig, ax = plt.subplots(figsize=(8,8))
explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, explode = explode, colors = colors, shadow = True, startangle = 90, labels = x_axis, autopct = '%.1f%%')
plt.gca().invert_xaxis()
plt.show()
```
